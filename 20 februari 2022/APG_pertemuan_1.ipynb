**Analisis tanpa outliers**

```r
Q <- quantile(dataset1$Jajan, probs=c(.25, .75), na.rm = FALSE)
iqr <- IQR(dataset1$Jajan)
eliminated<- subset(dataset1, dataset1$Jajan > (Q[1] - 1.5*iqr) & dataset1$Jajan < (Q[2]+1.5*iqr))
eliminated
```

```r
ggplot(eliminated, 
       aes(x = Jajan, 
           y = Jajan)) +
  geom_boxplot() +
  labs(title = "Salary distribution by rank")
```
